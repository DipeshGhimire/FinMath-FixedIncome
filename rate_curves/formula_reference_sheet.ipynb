## Formula Reference and Derivations

### 1. Spot Rate Conversions
We derive spot rates from the bootstrapped discount factor $Z(0,T)$.

**A. Semiannually Compounded Spot Rate ($r_{semi}$)**
* **Starting Formula:** Discounting $\$1$ over $T$ years with semi-annual compounding.
    $$Z(0,T) = \frac{1}{\left(1 + \frac{r_{semi}}{2}\right)^{2T}}$$
* **Derivation:**
    1. Invert: $\frac{1}{Z(0,T)} = \left(1 + \frac{r_{semi}}{2}\right)^{2T}$
    2. Take the $(2T)$-th root: $\left(\frac{1}{Z(0,T)}\right)^{\frac{1}{2T}} = 1 + \frac{r_{semi}}{2}$
    3. Solve for $r_{semi}$:
       $$r_{semi} = 2 \times \left( \left(\frac{1}{Z(0,T)}\right)^{\frac{1}{2T}} - 1 \right)$$

**B. Continuously Compounded Spot Rate ($r_{cont}$)**
* **Starting Formula:** Continuous discounting.
    $$Z(0,T) = e^{-r_{cont} \cdot T}$$
* **Derivation:**
    1. Take natural log ($\ln$): $\ln(Z(0,T)) = -r_{cont} \cdot T$
    2. Solve for $r_{cont}$:
       $$r_{cont} = -\frac{\ln(Z(0,T))}{T}$$

### 2. Forward Rate Conversions
First, we calculate the Forward Discount Factor $F$ for the interval $[T, T+0.5]$:
$$F(0, T, T+0.5) = \frac{Z(0, T+0.5)}{Z(0, T)}$$

**A. Semiannually Compounded Forward Rate ($f_{semi}$)**
* **Starting Formula:** The forward factor covers a time step of $\Delta t = 0.5$ years.
    $$F = \frac{1}{\left(1 + \frac{f_{semi}}{2}\right)^{2 \cdot \Delta t}}$$
* **Derivation:**
    1. Substitute $\Delta t = 0.5$ (so the exponent becomes $2 \times 0.5 = 1$):
       $$F = \frac{1}{1 + \frac{f_{semi}}{2}}$$
    2. Invert: $\frac{1}{F} = 1 + \frac{f_{semi}}{2}$
    3. Solve for $f_{semi}$:
       $$f_{semi} = 2 \times \left( \frac{1}{F} - 1 \right)$$

**B. Continuously Compounded Forward Rate ($f_{cont}$)**
* **Starting Formula:**
    $$F = e^{-f_{cont} \cdot \Delta t}$$
* **Derivation:**
    1. Take natural log ($\ln$): $\ln(F) = -f_{cont} \cdot \Delta t$
    2. Substitute $\Delta t = 0.5$:
       $$f_{cont} = -\frac{\ln(F)}{0.5}$$
